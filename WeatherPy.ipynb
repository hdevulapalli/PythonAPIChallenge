```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import urllib

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

#print(api_key)

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# Generate Cities List
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)


# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        #print(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
#print(cities)
```

```python
#Perform API Calls
#Perform a weather check on each city using a series of successive API calls.
#Include a print log of each city as it'sbeing processed (with the city number and city name).

#Base url for Weather Map API Call 
base_url = "http://api.openweathermap.org/data/2.5/forecast?id=524901&APPID=" + api_key

#City Data create a List
city_data=[]

#Print Logger 
print("Beginning Data Retrieval")     
print("-----------------------------")

#Counters for capturing all Counts
count_records = 1
count_set = 1

#Loop through the Cities List

for i, city in enumerate(cities):
    
    #Loop through Cities by setting them in a Group which contains Cities in Sets of 50
    if(i % 50 == 0 and i >= 50):
        #Set the Count of the Set to 1 and increment in sets
        count_set += 1
        #resetting record count to zero after Set is 1 - for a new set
        count_records=0
        
        
    #Create an Endpoint url for EACH CITY in the list
    city_url = base_url + "&q=" + urllib.request.pathname2url(city)
    
    #Start Logging the URL and set the Counts for Set and Record counter variables
    print(f"Processing Record {count_records} of Set {count_set} | {city}")
    #print(city_url)
    
    #Increase Counter for Records by 1
    count_records += 1
    
    #Perform a weather check on each city using a series of successive API calls.
    try:
        #Parse the JSON and retrieve the Data for each City
        city_weather = requests.get(city_url).json()
#         print(city_weather)
#         print (type(city_weather))
#         print(city_weather.keys())
#         print(type(city_weather['city']))
        #for cl in city_weather['city']:
            #print(cl.keys())
            #             print(type(cl))
            #             print(cl)
        cllist = pd.DataFrame(city_weather['list'])
        #print(cllist)
        maindf =cllist['main'].apply(pd.Series)
        #print(maindf['temp_max'].idxmax())
        weatherdf = cllist['weather'].apply(pd.Series)[0].apply(pd.Series)
        #print(cllist.iloc[maindf['temp_max'].idxmax()])
        #print(maindf.iloc[maindf['temp_max'].idxmax()])
        #print(weatherdf.iloc[maindf['temp_max'].idxmax()])
        #Run API Request for each City 
        #Parse all Weather information for that City
        city_lat = city_weather['city']["coord"]["lat"]
        city_lng = city_weather['city']["coord"]["lon"]
        city_max_temp = maindf.iloc[maindf['temp_max'].idxmax()]["temp_max"]
        city_humidity = maindf.iloc[maindf['temp_max'].idxmax()]["humidity"]
        city_clouds = cllist.iloc[maindf['temp_max'].idxmax()]["clouds"]["all"]
        city_wind = cllist.iloc[maindf['temp_max'].idxmax()]["wind"]["speed"]
        city_country = city_weather["city"]["country"]
        city_date = cllist.iloc[maindf['temp_max'].idxmax()]["dt"]
        
        #Store and add City information to the City Data List for each City
        city_data.append({"City": city,
                        "Cloudiness": city_clouds,
                        "Country": city_country,
                        "Date": city_date,
                        "Humidity": city_humidity,
                        "Lat": city_lat,
                        "Lng": city_lng,
                        "Max Temp": city_max_temp,
                        "Wind Speed": city_wind
                        })
        #print(city_data)
    except Exception as e:
        print(e)
        print("City not found. Skipping...")
        pass
    
print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")    
   
```

```python
#Convert City Data list into a Pandas Dataframe
#Convert Raw Data to DataFrame
city_data_pd = pd.DataFrame(city_data) 
# Display the DataFrame
city_data_pd.head()
```

```python
#Check what columns are present in this above DF
#city_data_pd.info()

#Extract Relevant Data
lats=city_data_pd["Lat"]
max_temps=city_data_pd["Max Temp"]
humidity=city_data_pd["Humidity"]
cloudiness = city_data_pd["Cloudiness"]
wind_speed = city_data_pd["Wind Speed"]
```

```python
#Export the city data into a csv
city_data_pd.to_csv(output_data_file, index_label="City_ID")
```
